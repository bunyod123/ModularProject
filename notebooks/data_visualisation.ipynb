```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\bunyo\OneDrive\Desktop\gitProjects\2_week_project\data\raw_data\LosAngeles_Earthquake_Dataset.csv")
```

```python
df.info()
```

```python
class_counts = df['class'].value_counts().sort_index()
plt.figure(figsize=(8, 6))
plt.bar(class_counts.index.astype(str), class_counts.values)
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Number of Earthquakes per Class')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.hist(df['rolling_mean_depth_30_days'], bins=30, edgecolor='black')
plt.xlabel('Rolling Mean Depth (km)')
plt.ylabel('Frequency')
plt.title('Zilzilalar chuqurligi vaqt bo‘yicha qanday o‘zgarayotgani')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['longitude'], df['latitude'], alpha=0.3, s=5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Qaysi joylarda zilzilalar ko‘proq sodir bo‘lgani')
plt.show()
```
